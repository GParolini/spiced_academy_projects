```python
import pandas as pd
import matplotlib.pyplot as plt

# Set the figure size to (14,6)
plt.rcParams['figure.figsize'] = (14,6)
```

# 1) Load the Data

```python
flights = pd.read_csv('flights_train.csv', index_col=0, parse_dates=True)
flights.head()
```

```python
# What datatype is the date?
flights.index
```

# 2) Plot the Data

```python
flights.plot()
plt.title('Monthly Passenger Numbers in 1000 over Time')
plt.ylabel('# of Passengers in 1000')
plt.ylim(ymin=0)
plt.show()
```
