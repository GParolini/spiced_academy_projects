```python
import pandas as pd
import matplotlib.pyplot as plt

# Set the figure size to (14,6)
plt.rcParams['figure.figsize'] = (14,6)
```

# 1) Load the Data

```python
flights = pd.read_csv('flights_train.csv', index_col=0, parse_dates=True)
flights.head()
```

```python
# What datatype is the date?
flights.index
```

# 2) Plot the Data

```python
flights.plot()
plt.title('Monthly Passenger Numbers in 1000 over Time')
plt.ylabel('# of Passengers in 1000')
plt.ylim(ymin=0)
plt.show()
```

## What do we see in the data?

- **Trend**; the number of passengers per month is upward sloping over time
- **Seasonality**; there is a periodically recurring pattern in the data; passenger numbers are higher in summer than in winter

Trend and seasonality are 2 of the 3 components I mentioned earlier. You will discuss how to model trend and seasonality in the afternoon.

The third component of time series data is the *remainder*. The remainder is the part of the time series data that can not be modelled by trend and seasonality.

The *remainder* itself consists of two components:

- time dependence (what we will focus on this week)
- randomness
