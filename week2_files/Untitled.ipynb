```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data/penguins_simple.csv", sep = ';')
```

```python
df
```

```python
df['Species'].value_counts()
color=['blue']*146 + ['pink']*119 + ['green']*68
df.plot.scatter(x='Culmen Length (mm)',
                y='Culmen Depth (mm)',
               c= color);
```

```python

df.plot.scatter(x='Culmen Length (mm)',
                y='Culmen Depth (mm)',
               c= pd.factorize(df['Species'])[0]);
```

```python
sns.scatterplot(data=df, x ='Culmen Length (mm)',
                y='Culmen Depth (mm)',
           hue ='Species');
```
