```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url_berlin = 'https://en.wikipedia.org/wiki/Berlin'
```

```python
response_berlin = requests.get(url_berlin)
```

```python
response_berlin.status_code
```

```python
html_berlin = response_berlin.text
```

```python
type(html_berlin)
```

```python
len(html_berlin)
```

```python
parsed_html_berlin = BeautifulSoup(html_berlin)
```

```python
label_requested = parsed_html_berlin.find_all('a', href=True)
```

```python
#label_requested
```

```python
type(label_requested)
```

```python
our_items = []
for item in label_requested:
    our_items.append(item.get("href"))
```
