## Imports and Loading Data

```python
import pandas as pd
import numpy as np

# NMF
from sklearn.decomposition import NMF
```

```python
df_ratings = pd.read_csv('data/ratings.csv')
df_ratings
```

```python
df_movies = pd.read_csv('data/movies.csv')
df_movies
```

```python
# Creating a dictionary of the movie titles and id
df_movies.drop(['genres'], axis=1, inplace=True)
movies_dict= dict(df_movies.values)
#movies_dict
```

## NMF

```python
df_ratings.drop('timestamp', axis=1, inplace=True)
R=df_ratings
```

```python
R.isnull().sum() #no missing values
```

```python
len(R)
```

```python
R_wide = pd.pivot_table(R, values='rating', index=['userId'], columns= ['movieId'])
```

```python
R_wide
```

```python
R_wide.isnull().sum() #several missing values
```

```python
R_wide.shape
```

```python
R_wide = R_wide.fillna(0)
```

```python
R_wide
```

```python
nmf = NMF(n_components=610)
```

```python
nmf.fit(R_wide)
```

```python
Q = nmf.components_
```

```python
Q.shape
```

```python
P = nmf.transform(R_wide)
```

```python
P.shape
```

```python
P = pd.DataFrame(nmf.transform(R_wide), index=R_wide.index)
```

```python
P
```

```python
R_wide_hat = pd.DataFrame(np.dot(P, Q), columns=R_wide.columns, index=R_wide.index)
round(R_wide_hat, 2)
```

```python
nmf.reconstruction_err_
```

## Prediction based on new user input

```python
# Create a dictionary for a new user
new_user_input = {'Toy Story (1995)': 4.0, 'Grumpier Old Men (1995)': 4.0}
new_user_input_keys = new_user_input.keys()

def get_key(val):
    for key, value in movies_dict.items():
         if val == value:
             return key

input_dict= {}
for key in new_user_input_keys:
    my_movieId = get_key(key)
    rating = new_user_input[key]
    input_dict.update({my_movieId:rating})
    
print(input_dict)
```

```python
# Convert it to a pd.DataFrame
new_user = pd.DataFrame(input_dict, index=['userXXX'], columns=R_wide.columns)
new_user
```

```python
#Prediction step 1 - generate user_P 
user_P = nmf.transform(new_user.fillna(0))
user_P
```

```python
#new user R - reconstruct R but for this new user only
user_R = pd.DataFrame(np.dot(user_P, Q), columns=R_wide.columns, index=['userXXX'])
user_R
```

```python
user_R.loc[:,list(set(user_R.columns).difference(set(new_user_input.keys())))]
```

```python
# Remove movies that have already been seen
recommendations = user_R.drop(columns=input_dict.keys())
recommendations
```

```python
# Sort the recomendations
my_rec = recommendations.sort_values(axis=1, by='userXXX', ascending=False)
my_rec
```

```python
rec_dict = my_rec.to_dict()
```

```python
rec_dict_5 = {k: rec_dict[k] for k in list(rec_dict)[0:5]}
```

```python
rec_dict_5
```

```python
my_rec_dict = {}
for key in rec_dict_5.keys():
    my_rec_dict.update({key:rec_dict_5[key]['userXXX'] })

user_rec = {}
for key in my_rec_dict:
    user_rec.update({movies_dict[key]:my_rec_dict[key]})

print(user_rec)
```
