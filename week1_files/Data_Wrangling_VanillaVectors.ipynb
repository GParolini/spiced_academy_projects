# 1. Loading and Combining Datasets

### 1.1 Load datasets
- One ``.csv`` file for each species: `penguins_gentoo.csv`, `penguins_adelie.csv`, `penguins_chinstrap.csv`
- How can we combine them all into one DataFrame?

```python
df1 = pd.read_csv("penguins_gentoo.csv")
df2 = pd.read_csv("penguins_adelie.csv")
df3 = pd.read_csv("penguins_chinstrap.csv")
```

```python
import os
```

```python
os.listdir()
```

```python
df_list = []
for file in os.listdir():
    if file.startswith('penguins'):
        df = pd.read_csv(file)
        df_list.append(df)
print(df_list)
```

```python
penguins_df = pd.concat(df_list) #column names must be the same; alt. append()
```

```python
penguins_df
```

### 1.2 Uh-oh! We're missing the sex data!
- Fortunately, that data exists in another `.csv` file: `sex_data_penguins.csv` 
- How can we "merge" the sex data into our current DataFrame?

```python
sex_df = pd.read_csv("sex_data_penguins.csv")
```

```python
sex_df
```

```python
penguins_df
```

```python
sex_df["Sample ID"] = sex_df['Species'].str[0] + "_" + sex_df['Sample Number'].astype(str)
```

```python
sex_df['Species'].str[0]
```

```python
sex_df.dtypes
```

```python
sex_df
```

```python
sex_df.drop(["Species", "Sample Number"], axis = 1) # or sex_df = sex_df(["Species", "Sample Number"], axis = 1)
```

```python
df_full = pd.merge(left=penguins_df, right=sex_df, how = 'left', on = "Sample ID")
df_full.head()
```

# 2. How to deal with missing values?

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_full = df_full.drop("Comments", axis=1)
```

```python

plt.figure(figsize=(12, 8))
sns.heatmap(df_full.isna(),cbar=False);
```

```python
df_full = df_full.dropna()
```

```python
sns.heatmap(df_full.isna(),cbar=False);
```
